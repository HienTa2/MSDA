#### Python Stats Methods


*    ttest_1samp: Conducts a one-sample T-test.
*    ttest_ind: Conducts a two-sample T-test for independent samples.
*    ttest_rel: Conducts a T-test on two related samples.
*    chi2_contingency: Conducts a Chi-squared test of independence for categorical data.
*    f_oneway: Performs a one-way ANOVA test.
*    pearsonr: Calculates a Pearson correlation coefficient.
*    spearmanr: Calculates a Spearman rank-order correlation coefficient.
*    kendalltau: Calculates Kendall's tau, a measure of ordinal association.
*    wilcoxon: Conducts the Wilcoxon signed-rank test.
*    mannwhitneyu: Performs the Mann-Whitney rank test on two independent samples.
*    kruskal: Conducts the Kruskal-Wallis H-test for independent samples.
*    friedmanchisquare: Conducts the Friedman test for repeated measurements.
*    anova_lm: Performs ANOVA (Analysis of Variance) using a linear model.
*    linregress: Performs a linear regression.
*    logistic: Performs logistic regression.
*    zscore: Computes the z score of each value in the sample, relative to the sample mean and standard deviation.
*    iqr: Calculates the interquartile range of the data.
*    skew: Measures the skewness of the data set.
*    kurtosis: Measures the kurtosis of the dataset.
*    norm: All functions related to the normal distribution.
*    expon: All functions related to the exponential distribution.
*    binom: All functions related to the binomial distribution.
*    poisson: All functions related to the Poisson distribution.



*    T-tests:
        ttest_ind: Independent two-sample T-test (assumes equal variances by default).
        ttest_ind(..., equal_var=False): Welch’s T-test, which does not assume equal population variance.
        ttest_rel: Paired T-test, for comparing two related samples.

*    Non-parametric Tests (when data don't follow a normal distribution or when sample sizes are small):
        mannwhitneyu: Mann-Whitney U test, a non-parametric test for assessing whether two independent samples come from the same distribution.
        wilcoxon: Wilcoxon signed-rank test, used for comparing two related samples to assess whether their population mean ranks differ.
        kruskal: Kruskal-Wallis H-test, a non-parametric method for comparing more than two independent samples.

*    ANOVA and Variants (for more than two groups, but can be adapted for two):
        f_oneway: One-way ANOVA, tests whether the means of two or more groups are different.
        anova_lm: ANOVA using a linear model, from the statsmodels library.

*    Correlation Tests (to measure the strength and direction of association between two numeric variables):
        pearsonr: Pearson correlation coefficient, for linear relationship between datasets.
        spearmanr: Spearman rank-order correlation coefficient, for non-linear relationships.
        kendalltau: Kendall’s tau, another non-parametric measure of relationships between columns of ranked data.

*    Equivalence and Non-Inferiority Tests (to determine if two groups are statistically equivalent or one is not worse than the other):
        These tests are more specialized and often require specific statistical software or packages that support such analyses.

#### Different Stats Testing

*    Descriptive Statistics: Methods to summarize data characteristics, like mean, median, mode, variance, and standard deviation.

*    Inferential Statistics: Techniques to make inferences about populations based on samples, which include:
        Hypothesis Testing: Determining if there is enough evidence to reject a null hypothesis, like with T-tests (one-sample, independent two-sample, paired).
        Confidence Intervals: Estimating intervals within which population parameters likely lie, with a given level of confidence.

*    Correlation and Covariance: Measuring the relationship and dependence between two variables.

*   Regression Analysis: Understanding how the typical value of the dependent variable changes when any one of the independent variables is varied.

*    ANOVA (Analysis of Variance): Comparing the means of three or more groups to see if at least one mean is different from the others.

*   Chi-Square Test: Assessing relationships between categorical variables to see if distributions of categorical variables differ from each other.

*   Non-parametric Tests: Used when data don't fit the normal distribution or when the sample size is too small, like the Mann-Whitney U test or the Kruskal-Wallis test.

*    Time Series Analysis: Techniques to analyze time-ordered data points to extract meaningful statistics and characteristics.

*    Factor Analysis: Identifying underlying variables, or factors, that explain the pattern of correlations within a set of observed variables.
*    Cluster Analysis: Grouping a set of objects in such a way that objects in the same group are more similar to each other than to those in other groups.

*    Survival Analysis: Analyzing the expected duration of time until one or more events happen, like death in biological organisms and failure in mechanical systems.
